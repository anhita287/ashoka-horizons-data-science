```python
!pip install pandas
!pip install numpy
!pip install seaborn
!pip install matplotlib
import pandas as pd

df= pd.read_csv('diabetes.csv')

print('Top 5 entries:')
print(df.head())

print('Bottom 5 entries:')
print(df.tail())

print (df.dtypes)

import seaborn as sns
import matplotlib.pyplot as plt

#Question: How does BMI affect the likelihood of diabetes?
#Question: Is there a correlation between the number of pregnancies and diabetes?

#Distribution of BMI
sns.histplot(df['BMI'], kde=True)
plt.title('Distribution of BMI')
plt.show()

# Relationship between BMI and Outcome
sns.boxplot(x='Outcome', y='BMI', data=df)
plt.title('BMI vs Outcome')
plt.show()

# Correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
